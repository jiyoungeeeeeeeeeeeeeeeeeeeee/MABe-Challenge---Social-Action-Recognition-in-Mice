```python
# mouse1 -> 1 = Ï†ïÏàò Î≥ÄÌôò Ìï®Ïàò
def to_id(values):
  if isinstance(values , str) and values.startswith("mouse"):
    return int(values.replace('mouse' , ''))
  return values
```

```python
import ast
# meta_data ÎßåÎì§Í∏∞

def get_agent_target_action (meta_train , video_id):

  df = meta_train.copy()
  df['video_id'] = df['video_id'].astype(str).str.strip()
  vid = str(video_id).strip()

  mask = df['video_id'] == vid
  row  = df.loc[mask].iloc[0]

  behaviors = ast.literal_eval(row['behaviors_labeled'])

  s = behaviors[0].split(",")

  fps  = float(row['frames_per_second'])
  ppcm = float(row['pix_per_cm_approx'])


  return  {
      'video_id': video_id ,
      'agent' : to_id(s[0]),
      'target': to_id(s[1]),
      'action':      (s[2]),
      'fps'   : fps,
      'ppcm'  : ppcm
  }
```

```python
meta_data = get_agent_target_action(meta_train = meta_train , video_id='44566106')
meta_data
```

```python
def pix_exchange_cm(trk , meta_data):

  use = trk[['video_frame','mouse_id','bodypart','x','y']].copy()
  use[['x','y']] = use[['x','y']]/meta_data['ppcm']

  # ÌîÑÎ†àÏûÑ √ó mouse_id Î≥Ñ ÌèâÍ∑† Ï¢åÌëú Í≥ÑÏÇ∞
  cen = (use.groupby(["video_frame","mouse_id"], as_index=False)[["x","y"]]
           .mean().rename(columns={"x":"cx","y":"cy"}))

  return cen
```

```python
tkr_cem = pix_exchange_cm(trk= trk , meta_data= meta_data)
tkr_cem.head(3)
```

```python
import numpy as np

def speed (cen_df,fps):

  cen_sorted = cen_df.sort_values(['mouse_id','video_frame']).copy()

  cen_sorted[['dcx' , 'dcy']] = cen_sorted.groupby("mouse_id")[['cx' , 'cy']].diff()
  cen_sorted['speed'] = (np.hypot(cen_sorted['dcx'],cen_sorted['dcy']) * fps ).fillna(0)

  return cen_sorted[['video_frame', 'mouse_id', 'speed' , 'cx' , 'cy']]
```

```python
trk_cen_speed = speed(tkr_cem , fps= meta_data['fps'])
trk_cen_speed.head(3)
```

```python
def relat_agent_target(cen_speed_df, fps, agent_id, target_id ):

  A = cen_speed_df[cen_speed_df.mouse_id== agent_id].rename(columns={"cx":"ax","cy":"ay","speed":"aspeed"})

  B = cen_speed_df[cen_speed_df.mouse_id== target_id].rename(columns={"cx":"bx","cy":"by","speed":"bspeed"})

  G = A.merge(B, on="video_frame", how="inner").sort_values('video_frame')

  G["dist"]     = np.hypot(G["ax"]-G["bx"], G["ay"]-G["by"])

  G["approach"] = G["dist"].diff().fillna(0) * fps   # ÏùåÏàòÎ©¥ Í∞ÄÍπåÏõåÏßê(Ï∂îÍ≤© ÏÑ±Ìñ•)

  return A,B,G
```

```python
agent_id = meta_data['agent']
target_id = meta_data['target']
fps = meta_data['fps']

A, B, G = relat_agent_target(trk_cen_speed, fps, agent_id, target_id)

G.head()
```

## 3) ÏúàÎèÑ X + y(ÎùºÎ≤®) ÎßåÎì§Í∏∞

```python
# Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖòÏóêÏÑú ÏúÑÏóêÏÑú Ï†ïÏùòÌïú Î©îÌÉÄ Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞

def mask_pos_Xy(actions):

  mask = (ann.action == actions) & (ann.agent_id== agent_id) & (ann.target_id==target_id)
  pos  = ann[mask][['start_frame','stop_frame']].to_numpy()

  return pos
```

```python
meta_Xy = mask_pos_Xy(meta_data['action'])
meta_Xy
```

```python
def make_windows_for_pair(G, pos_segments, W=60, S=15,
                          video_id=None, agent=None, target=None):
    """
    G: one pair dataframe (columns: video_frame, aspeed, bspeed, dist, approach, ...)
    pos_segments: [(start_frame, stop_frame), ...]  # Ìï¥Îãπ ÌñâÎèô Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖò Íµ¨Í∞Ñ
    """
    # 0) ÎπÑÏñ¥ÏûàÎäî Ïåç Îπ†Î•¥Í≤å Ïä§ÌÇµ
    if G is None or len(G)==0 or "video_frame" not in G:
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])

    # 1) video_frame ÏµúÎåÄÍ∞í ÏïàÏ†Ñ Í≥ÑÏÇ∞
    max_frame = pd.to_numeric(G["video_frame"], errors="coerce").max()
    if pd.isna(max_frame):
        # ÌîÑÎ†àÏûÑ Ï†ïÎ≥¥Í∞Ä ÏóÜÏúºÎ©¥ Ïä§ÌÇµ
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])
    n_frames = int(max_frame) + 1

    # 2) ÌîÑÎ†àÏûÑ ÎùºÎ≤®(y) ÎßåÎì§Í∏∞ (Íµ¨Í∞ÑÏù¥ ÏóÜÏúºÎ©¥ Ï†ÑÎ∂Ä 0)
    y = np.zeros(n_frames, dtype=np.int8)
    if pos_segments is not None and len(pos_segments) > 0:
        for s,e in pos_segments:
            s = max(0, int(s))
            e = min(n_frames-1, int(e))
            if s <= e:
                y[s:e+1] = 1

    # 3) ÏúàÎèÑ ÏöîÏïΩ (Í∏∏Ïù¥ Î∂ÄÏ°±ÌïòÎ©¥ Îπà DF)
    vf = G["video_frame"].to_numpy()
    if len(vf) < W:
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])

    rows = []
    for s_idx in range(0, len(vf)-W+1, S):
        w = G.iloc[s_idx:s_idx+W]
        s, e = int(w.video_frame.iloc[0]), int(w.video_frame.iloc[-1])
        # y ÏúàÎèÑ ÎùºÎ≤® (Í≤ΩÍ≥Ñ ÏïàÏ†Ñ Ïä¨ÎùºÏù¥Ïä§)
        ys = y[s:e+1] if e < len(y) else y[s:]
        y_win = int(ys.mean() >= 0.5) if len(ys) > 0 else 0

        rows.append({
            "video_id":   video_id,
            "agent_id":   agent,
            "target_id":  target,
            "start_frame": s,
            "end_frame":   e,
            "dist_mean":     float(w["dist"].mean()),
            "dist_std":      float(w["dist"].std(ddof=0)),
            "aspeed_mean":   float(w["aspeed"].mean()),
            "bspeed_mean":   float(w["bspeed"].mean()),
            "approach_mean": float(w["approach"].mean()),
            "y": y_win,
        })
    return pd.DataFrame(rows)
```

```python
Xy = make_windows_for_pair(
    G,
    pos_segments=meta_Xy, W=W, S=S,
    video_id= meta_data['video_id'], agent=agent_id, target=target_id
)
Xy.head(3)
```
