```python
# mouse1 -> 1 = Ï†ïÏàò Î≥ÄÌôò Ìï®Ïàò
import re

def to_id(x):
  _mouse_pat = re.compile(r"""^["']?\s*mouse\s*(\d+)\s*["']?$""", re.IGNORECASE)
  if isinstance(x, str):
      s = x.strip(" '\"")          # ÏïûÎí§ Í≥µÎ∞±/Îî∞Ïò¥Ìëú Ï†úÍ±∞
      m = _mouse_pat.match(s)
      if m:
          return int(m.group(1))   # mouseN ‚Üí N
      return s
  return x
```

```python
import ast

def behavior_token(df):

  behavior_labe = df['behaviors_labeled'].fillna('[]')
  result = []

  for labe in behavior_labe:
    behavior = ast.literal_eval(labe)
    token = []

    for b in behavior:
      parts = [to_id(x.strip()) for x in b.split(',')]
      token.append(parts)

    result.append(token)

  return result
```

```python
# meta_data ÎßåÎì§Í∏∞

def get_agent_target_action (df, video_id):

  df = meta_train.copy()
  df['video_id'] = df['video_id'].astype(str).str.strip()
  vid = str(video_id).strip()

  mask = df['video_id'] == vid
  row  = df.loc[mask].iloc[0]

  fps  = float(row['frames_per_second'])
  ppcm = float(row['pix_per_cm_approx'])


  # agent - target



  return  {
      'video_id': video_id ,
      # "agent":  ,
      # "target": ,
      # "action": ,
      'fps'   : fps,
      'ppcm'  : ppcm
  }
```

```python
def pix_exchange_cm(trk , meta_data):

  use = trk[['video_frame','mouse_id','bodypart','x','y']].copy()
  use[['x','y']] = use[['x','y']]/meta_data['ppcm']

  # ÌîÑÎ†àÏûÑ √ó mouse_id Î≥Ñ ÌèâÍ∑† Ï¢åÌëú Í≥ÑÏÇ∞
  cen = (use.groupby(["video_frame","mouse_id"], as_index=False)[["x","y"]]
           .mean().rename(columns={"x":"cx","y":"cy"}))

  return cen
```

```python
tkr_cem = pix_exchange_cm(trk= trk , meta_data= meta_data)
tkr_cem.head(3)
```

```python
import numpy as np

def speed (cen_df,fps):

  cen_sorted = cen_df.sort_values(['mouse_id','video_frame']).copy()

  cen_sorted[['dcx' , 'dcy']] = cen_sorted.groupby("mouse_id")[['cx' , 'cy']].diff()
  cen_sorted['speed'] = (np.hypot(cen_sorted['dcx'],cen_sorted['dcy']) * fps ).fillna(0)

  return cen_sorted[['video_frame', 'mouse_id', 'speed' , 'cx' , 'cy']]
```

```python
trk_cen_speed = speed(tkr_cem , fps= meta_data['fps'])
trk_cen_speed.head(3)
```

```python
def relat_agent_target(cen_speed_df, fps, agent_id, target_id ):

  A = cen_speed_df[cen_speed_df.mouse_id== agent_id].rename(columns={"cx":"ax","cy":"ay","speed":"aspeed"})

  B = cen_speed_df[cen_speed_df.mouse_id== target_id].rename(columns={"cx":"bx","cy":"by","speed":"bspeed"})

  G = A.merge(B, on="video_frame", how="inner").sort_values('video_frame')

  G["dist"]     = np.hypot(G["ax"]-G["bx"], G["ay"]-G["by"])

  G["approach"] = G["dist"].diff().fillna(0) * fps   # ÏùåÏàòÎ©¥ Í∞ÄÍπåÏõåÏßê(Ï∂îÍ≤© ÏÑ±Ìñ•)

  return A,B,G
```

## 3) ÏúàÎèÑ X + y(ÎùºÎ≤®) ÎßåÎì§Í∏∞

```python
# Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖòÏóêÏÑú ÏúÑÏóêÏÑú Ï†ïÏùòÌïú Î©îÌÉÄ Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞

def mask_pos_Xy(actions):

  mask = (ann.action == actions) & (ann.agent_id== agent_id) & (ann.target_id==target_id)
  pos  = ann[mask][['start_frame','stop_frame']].to_numpy()

  return pos
```

```python
def make_windows_for_pair(G, pos_segments, W=60, S=15,
                          video_id=None, agent=None, target=None):
    """
    G: one pair dataframe (columns: video_frame, aspeed, bspeed, dist, approach, ...)
    pos_segments: [(start_frame, stop_frame), ...]  # Ìï¥Îãπ ÌñâÎèô Ïñ¥ÎÖ∏ÌÖåÏù¥ÏÖò Íµ¨Í∞Ñ
    """
    # 0) ÎπÑÏñ¥ÏûàÎäî Ïåç Îπ†Î•¥Í≤å Ïä§ÌÇµ
    if G is None or len(G)==0 or "video_frame" not in G:
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])

    # 1) video_frame ÏµúÎåÄÍ∞í ÏïàÏ†Ñ Í≥ÑÏÇ∞
    max_frame = pd.to_numeric(G["video_frame"], errors="coerce").max()
    if pd.isna(max_frame):
        # ÌîÑÎ†àÏûÑ Ï†ïÎ≥¥Í∞Ä ÏóÜÏúºÎ©¥ Ïä§ÌÇµ
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])
    n_frames = int(max_frame) + 1

    # 2) ÌîÑÎ†àÏûÑ ÎùºÎ≤®(y) ÎßåÎì§Í∏∞ (Íµ¨Í∞ÑÏù¥ ÏóÜÏúºÎ©¥ Ï†ÑÎ∂Ä 0)
    y = np.zeros(n_frames, dtype=np.int8)
    if pos_segments is not None and len(pos_segments) > 0:
        for s,e in pos_segments:
            s = max(0, int(s))
            e = min(n_frames-1, int(e))
            if s <= e:
                y[s:e+1] = 1

    # 3) ÏúàÎèÑ ÏöîÏïΩ (Í∏∏Ïù¥ Î∂ÄÏ°±ÌïòÎ©¥ Îπà DF)
    vf = G["video_frame"].to_numpy()
    if len(vf) < W:
        return pd.DataFrame(columns=[
            "video_id","agent_id","target_id","start_frame","end_frame",
            "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
        ])

    rows = []
    for s_idx in range(0, len(vf)-W+1, S):
        w = G.iloc[s_idx:s_idx+W]
        s, e = int(w.video_frame.iloc[0]), int(w.video_frame.iloc[-1])
        # y ÏúàÎèÑ ÎùºÎ≤® (Í≤ΩÍ≥Ñ ÏïàÏ†Ñ Ïä¨ÎùºÏù¥Ïä§)
        ys = y[s:e+1] if e < len(y) else y[s:]
        y_win = int(ys.mean() >= 0.5) if len(ys) > 0 else 0

        rows.append({
            "video_id":   video_id,
            "agent_id":   agent,
            "target_id":  target,
            "start_frame": s,
            "end_frame":   e,
            "dist_mean":     float(w["dist"].mean()),
            "dist_std":      float(w["dist"].std(ddof=0)),
            "aspeed_mean":   float(w["aspeed"].mean()),
            "bspeed_mean":   float(w["bspeed"].mean()),
            "approach_mean": float(w["approach"].mean()),
            "y": y_win,
        })
    return pd.DataFrame(rows)
```

# Ï†ÑÏ≤òÎ¶¨ ÌååÏù¥ÌîÑÎùºÏù∏

```python
# wrapper Ìï®Ïàò
from pathlib import Path

def pipeline_build_video(meta_train , BASE , video_id , W = 60 , S = 15 ):

  info = get_agent_target_action( meta_train = meta_train, video_id = video_id)

  trk  = load_tracking_for_video(BASE , info)
  ann  = load_train_annotation(BASE , info)

  tkr_cem        = pix_exchange_cm(trk = trk , meta_data = info)
  trk_cen_speed  = speed(tkr_cem , fps = info['fps'])

  agent_id  = info['agent']
  target_id = info['target']
  fps       = info['fps']

  A, B, G = relat_agent_target(trk_cen_speed, fps, agent_id, target_id)

  meta_Xy = mask_pos_Xy(info['action'])

  Xy = make_windows_for_pair(
    G,
    pos_segments=meta_Xy, W=W, S=S,
    video_id= info['video_id'], agent=agent_id, target=target_id
)

  return Xy
```

```python
def iter_Xy(meta_train, BASE, W=60 , S=15):

  for vid in meta_train['video_id'].astype(str):
    try:
      Xy = pipeline_build_video(meta_train, BASE, video_id= vid, W=W, S=S)
      yield Xy

    except Exception as e:
      print(f"vid Ï≤òÎ¶¨ Ïã§Ìå® {e}")
      continue
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
from pathlib import Path

# Ïä§ÌÇ§Îßà Í∏∞Ï§ÄÏù¥ Îê† Ïª¨Îüº ÏàúÏÑúÏôÄ dtype ÏßÄÏ†ï
COLS = [
    "video_id","agent_id","target_id","start_frame","end_frame",
    "dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean","y"
]

def _normalize_df(df):
    # Ïó¥ ÏàúÏÑú Í≥†Ï†ï
    df = df.reindex(columns=COLS)
    # dtype Í≥†Ï†ï (nullable int ÏÇ¨Ïö©)
    df["video_id"] = df["video_id"].astype("string")
    for c in ["agent_id","target_id","start_frame","end_frame","y"]:
        df[c] = df[c].astype("Int64")
    for c in ["dist_mean","dist_std","aspeed_mean","bspeed_mean","approach_mean"]:
        df[c] = df[c].astype("float64")
    return df

def save_to_single_parquet(meta_train, BASE, out_path, W=60, S=15):
    out_path = Path(out_path)
    out_path.parent.mkdir(parents=True, exist_ok=True)

    writer = None
    first_schema = None
    wrote_any = False

    try:
        for Xy in iter_Xy(meta_train, BASE, W=W, S=S):
            if Xy is None or Xy.empty:
                continue

            # Ïä§ÌÇ§Îßà ÌùîÎì§Î¶º Î∞©ÏßÄ
            Xy = _normalize_df(Xy)

            table = pa.Table.from_pandas(Xy, preserve_index=False)

            if first_schema is None:
                first_schema = table.schema

            if writer is None:
                writer = pq.ParquetWriter(
                    out_path,
                    first_schema,
                    compression="zstd"
                )

            # Ïä§ÌÇ§ÎßàÍ∞Ä Îã§Î•¥Î©¥ Í∞ïÏ†ú Ï∫êÏä§ÌåÖ (ÏïàÏ†Ñ Î™®Îìú off)
            if table.schema != first_schema:
                table = table.cast(first_schema, safe=False)

            writer.write_table(table)
            wrote_any = True

    finally:
        if writer is not None:
            writer.close()

    if not wrote_any:
        print(" Ï†ÄÏû•Ìï† Ïú†Ìö®Ìïú Î∞∞ÏπòÍ∞Ä ÏóÜÏñ¥ ÌååÏùºÏùÑ ÏÉùÏÑ±ÌïòÏßÄ ÏïäÏïòÏäµÎãàÎã§.")
```

```python
out_path = BASE/'xy_full.parquet'
save_to_single_parquet(meta_train, BASE, out_path, W=W, S=S)
```
